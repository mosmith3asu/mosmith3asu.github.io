For my portion of the Dynamics II assignment, my objective was to create a physical prototype to determine the validity of our model and if it is feasible to assemble.

The team agreed that for our final system, most joints would consist of 0.005” thick Grafix Dura-Lar (a polyester film) sandwiched in between two pieces of 3D printed PLA. For the damping joint, 0.028” thick spring steel and a damping material would be sandwiched in between 3D printed PLA plates. For the damping joint, 3 damping materials planned to be tested, they are high density polyurethane foam, rubber, and polycarbonate. All damping materials will be 1/32” thick. Based on experimental results, the final damping material will be chosen for the system. A control model without the damping material and just the spring steel will also be tested.

Due to lead time on those materials, I was unable to build this initial prototype with those materials, with the exception of the 3D printed PLA. In place of the 0.005” Grafix Dura-Lar, I used plastic sheet from a folder sleeve. While I do not know the exact material, I suspect it is PET. I measured it to have a thickness of 0.007”, similar to the desired thickness of the Grafix Dura-Lar. The folder sleeve is pictured below in figure 1:


1.PNG

The material could be folded and retain it’s fold, but was also somewhat elastic in that it wanted to move back to a point of equilibrium.

The damping material used was 1/32” rubber splicing tape, this will likely have similar properties to the 1/32” rubber sheet the team has purchased. It has the added benefit of including adhesive on one side for ease of assembly.

For the 3D printed PLA parts, the link lengths were based on the model the team developed in the system kinematics assignment. For links that had a joint at either end, the length was reduced by 3mm in order to allow the plastic layer to bend and flex. For links that only had one joint at the end, their length was reduced by 1.5mm. Figure 2 shows a diagram from the system kinematics assignment with the original link lengths:


2.PNG

The width of each joint was agreed on by the team to be 30mm. In the team’s models, the current assumption is that the links are rigid. To ensure this is the case, the thickness of each PLA plate is currently 1.5mm. While one plate can be flexed by hand, the forces this system will experience will likely not cause it to flex. This is tested by FEA in the future.

Because there are some joints in the middle of a link, some additional pieces were designed such that there would be 3mm of space around the joint to allow the plastic layer to bend and join up with the link. All the pieces used can be seen in figure 3 below:

For the plastic layer, 3 pieces were cut from the folder and folded at each joint location. For the final build, this will be reduced to 2 pieces that will consist of one main loop and one small piece to join the front four bar mechanism. Figure 4 shows the plastic layers folded to form the shape of the prototype:

Pieces 1 and 2 will be combined for the final system. Piece 3 will also have more overlap onto piece 1 for the final system. The rubber splicing tape was added at the joint that connects the foot to the ankle tendon. The 3D printed parts were then super glued onto the plaster layers and connected into one system. The final prototype has a mass of 17 grams, and is able to stand under its own weight. With the correct initial link positions, the model is able to achieve a stable landing when dropped straight down. This prototype proves the validity of the team’s approach and will be refined in future assignments.There are three primary changes to make for the final system excluding assembly with the actual materials. The first is to only have two pieces for the plastic layer (as opposed to 3), and potentially doubling the thickness of the plastic layer. Doubling the thickness will depend on the rigidity of Grafix Dura-Lar. Right now, the joints are able flex a bit too much upon impact with the ground. The third change will be to replace the top 3D printed plate with a platform to allow a payload to be placed on the robot to replicate the actual mass of a kangaroo rat (or equivalent scale).
