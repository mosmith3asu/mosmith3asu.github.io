The work that I have done above will be used to find the forces experienced during impact  and the impulse experienced by the robot. From the video analysis, measure the acceleration at the top most link and find the force at every timestep after impact by multiplying the systems mass with the simple force equation 

$$F=ma$$.

This calculation will provide the force experienced by the robot at every time during impact. The impulse experienced by the robot can then be found by calculating the area under the force-time curve which is described by

$$\int_{t_1}^{t_1} F \, dt = \int_{t_1}^{t_1} m a(t) \, dt$$

## Part II

**Assumptions**

1. The friction between prototype and glass/wall during fall is negligible
2. Aligning the top link of the prototype with the glass will create reliable conditions for dropping the device from the same orientation and height


**Data collection Procedure:**

1. Begin video recording
2. Align the top most link of the prototype such that it is parallel to the top of the glass plate. 
3. Press a flat object against the back of the leg to hold in in this position
4. Quickly remove the flat object to drop the leg down between the pane of glass and the wall
5. Wait for the leg to come to rest (no bouncing)
6. Stop the video recording


**Tracker Video Analysis Procedure**


1.   Upload the video into the tracker software
2. Create an coordinate axis along the impact plane
3. Create a calibration stick on the measured grid
4. Track all marked points on the robot
5. Export the data as a CSV
6. Calculate the force over time impulse parameters according to equations 1 and 2
